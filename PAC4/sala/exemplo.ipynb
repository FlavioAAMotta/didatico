```python
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Definindo listas de exemplos
transportes = ['Ônibus Escolar', 'Onibus Prefeitura', 'Carro', 'Bicicleta', 'Moto']
trabalho = ['Sim', 'Não']

# Função para gerar 40 linhas de dados simulados
def gerar_dados_exemplo(num_linhas=50):
    dados = []
    for _ in range(num_linhas):
        idade = random.randint(18, 25)  # Idade entre 18 e 25 anos
        tempo_trajeto = random.randint(5, 90)  # Tempo de trajeto entre 5 e 90 minutos
        irmaos = random.randint(0, 5)  # Número de irmãos entre 0 e 5
        dependencias = random.randint(0, 3)  # Quantidade de dependências entre 0 e 3
        horas_estudo = random.randint(5, 20)  # Horas de estudo por semana entre 5 e 20
        transporte = random.choice(transportes)  # Escolhendo um transporte aleatório
        nota_media = round(random.uniform(5.0, 10.0), 1)  # Nota média entre 5.0 e 10.0
        horas_sono = round(random.uniform(5.0, 9.0), 1)  # Horas de sono entre 5.0 e 9.0
        trabalha = random.choice(trabalho)  # Trabalha ou não
        dados.append([idade, tempo_trajeto, irmaos, dependencias, horas_estudo, transporte, nota_media, horas_sono, trabalha])
    
    return dados

# Gerando os dados
dados_exemplo = gerar_dados_exemplo()

# Criando um DataFrame com os dados simulados
colunas = ['Idade', 'Tempo p chegar na faculdade (min)', 'Irmãos', 'Deps', 'Horas de Estudo/Semana', 'Transporte', 'Nota Média no Último Semestre', 'Horas de Sono por Noite', 'Trabalha']
df_exemplo = pd.DataFrame(dados_exemplo, columns=colunas)

# Exibindo os dados gerados
```
