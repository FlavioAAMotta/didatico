```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset Iris
df = pd.read_csv('doacoes-saida.csv', sep=';')
```

```python
df.head()
```

```python
df.loc[df['CIDADE_DESTINATARIO'] == "MURIAE"]
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Boxplot para analisar o comprimento da pétala por espécie
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Comprimento da Pétala por Espécie')
plt.show()
```
