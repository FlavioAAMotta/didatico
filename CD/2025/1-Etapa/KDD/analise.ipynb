```python
# Bloco 4.2: Identificar e Corrigir Salários Negativos
# Primeiro, verifique se há salários negativos:
print("Salários negativos antes da correção:")
display(df[df['Salario'] < 0])
```

```python
# # Podemos corrijir os salários negativos transformando-os em valores absolutos (não há garantia de ser esse o problema).
# df['Salario'] = df['Salario'].apply(lambda x: abs(x) if pd.notnull(x) else x)

# # Ou podemos substituir os salários negativos por NaN, se for o caso.
# df['Salario'] = df['Salario'].apply(lambda x: x if x >= 0 else pd.NaN)

# # Ou ainda, podemos substituir os salários negativos pela média dos salários positivos.
# # Primeiro, calculamos a média dos salários positivos:
# media_salarios = df[df['Salario'] >= 0]['Salario'].mean()
# # Substituímos os salários negativos pela média:
# df['Salario'] = df['Salario'].apply(lambda x: x if x >= 0 else media_salarios)

# # Ou podemos substituir os salários negativos pela mediana dos salários positivos por cargo.
# Primeiro, calcular a mediana dos salários positivos para cada cargo
mediana_por_cargo = df[df['Salario'] >= 0].groupby('Cargo')['Salario'].median().to_dict()

# Agora, substituímos os salários negativos pela mediana do cargo correspondente
df['Salario'] = df.apply(
    lambda row: row['Salario'] if row['Salario'] >= 0 else mediana_por_cargo.get(row['Cargo'], row['Salario']),
    axis=1
)

# # Ou podemos remover as linhas com salários negativos:
# df = df[df['Salario'] >= 0]

# Confira os primeiros valores corrigidos:
print("Salários após a correção:")
display(df.head(10))
```

```python
# Bloco 4.3: Padronizar Nomes dos Cargos
# Muitos cargos podem ter sido registrados com erros de digitação.
# Vamos criar uma função para mapear esses erros para a forma correta.
def padronizar_cargo(cargo):
    mapping = {
        'Desenolvedor': 'Desenvolvedor',
        'Dev': 'Desenvolvedor',
        'Programador': 'Desenvolvedor',
        'Estag': 'Estagiário',
        'Estagiario': 'Estagiário',
        'Trainee': 'Estagiário',
        'Gerent': 'Gerente',
        'Manager': 'Gerente',
        'Gerente de Projetos': 'Gerente',
        'Direktor': 'Diretor',
        'Director': 'Diretor',
        'Diret': 'Diretor',
        'Analist': 'Analista',
        'Analysta': 'Analista',
        'Analista de Sistemas': 'Analista',
        'Coord': 'Coordenador',
        'Coordenator': 'Coordenador',
        'Coordenaddr': 'Coordenador',
        'Auxilir': 'Auxiliar',
        'Assist': 'Auxiliar',
        'Helper': 'Auxiliar'
    }
    if pd.isnull(cargo):
        return cargo
    return mapping.get(cargo, cargo)

# Aplique a função para padronizar os cargos:
df['Cargo'] = df['Cargo'].apply(padronizar_cargo)

# Confira os cargos após a padronização:
print("Cargos após a padronização:")
display(df['Cargo'].value_counts())
```
