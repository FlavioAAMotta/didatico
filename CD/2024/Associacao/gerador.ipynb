```python
num_itens = 10000
num_itens_por_compra = np.random.normal(loc=5, scale=3.5, size=num_itens).astype(int)

# Limitando o número de itens por compra entre 2 e 20
num_itens_por_compra = np.clip(num_itens_por_compra, 2, 20)


num_itens_por_compra = np.clip(num_itens_por_compra, 1, 20)
nomes_completos = [np.random.choice(nomes) + ' ' + np.random.choice(sobrenomes) for _ in range(num_itens//10)]
```

```python
linhas = []

for num_itens in num_itens_por_compra:
    compra = []
    for _ in range(num_itens):
        item_escolhido = np.random.choice(itens)
        compra.append(item_escolhido)
        if item_escolhido in itens_relacionados:
            itens_rel = itens_relacionados[item_escolhido]
            for item_rel in itens_rel:
                if (np.random.rand() < 0.75) and (item_rel not in compra) and (len(compra) < 20):
                            compra.append(item_rel)
    # Garantir que a compra não tenha mais do que o número máximo de itens definido
    compra = compra[:num_itens]
    linhas.append(compra)

# Converter a lista de compras em um DataFrame
df_compras = pd.DataFrame(linhas)

# Adicionar uma coluna com o nome do cliente como primeia coluna
df_compras['cliente'] = np.random.choice(nomes_completos, size=len(df_compras))

# Reordenar as colunas para que o nome do cliente fique na primeira coluna
df_compras = df_compras[['cliente'] + list(range(df_compras.shape[1]-1))]

# Mostrar um exemplo das compras geradas
df_compras.head()
```

```python
transactions = []
for i in range(0,df_compras.shape[0]):
    transactions.append([str(df_compras.values[i,j])for j in range(0,df_compras.shape[1])])
```

```python
from apyori import apriori

regras = apriori(transactions=transactions, min_support=0.03, min_confidence=0.2, min_lift=3, min_length=2, max_length=4)
```

```python
# usando fp-growth
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

from mlxtend.preprocessing import TransactionEncoder

# Supondo que df_compras seja uma lista de listas, onde cada sublista é uma transação
# Se df_compras já é um DataFrame com cada item em uma coluna, ajuste conforme necessário
# Convertendo cada linha do DataFrame em uma lista, excluindo valores NaN
df_compras_lista = df_compras.apply(lambda x: x.dropna().tolist(), axis=1).tolist()

# Instancia um TransactionEncoder e ajusta aos dados
te = TransactionEncoder()
te_ary = te.fit(df_compras_lista).transform(df_compras_lista)

# Converte o array resultante em um DataFrame pandas
df_onehot = pd.DataFrame(te_ary, columns=te.columns_)

# Agora você pode aplicar o fpgrowth no DataFrame one-hot
frequent_itemsets = fpgrowth(df_onehot, min_support=0.03, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```
