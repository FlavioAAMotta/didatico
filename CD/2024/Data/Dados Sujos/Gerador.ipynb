```python
import numpy as np
import pandas as pd
```

```python
# Gerador de dataframe com dados aleatórios, considerando nome, nascimento, salário e cargo
def gera_dataframe(n):
    np.random.seed(0)
    nomes = ['ALYSON', 'DANIEL', 'DAVI', 'FILIPE', 'GABRIEL', 'GUSTAVO', 'JOÃO', 'JÚLIO', 'LUIZ', 'MAYCON', 'OTÁVIO', 'PEDRO', 'SERGIO', 'SÁVIO', 'VANESSA', 'VICTOR', 'WEMERSON']
    sobrenomes = ['BANDEIRA', 'CARDOSO', 'COSTA', 'CRUZ', 'DEMARQUE', 'LIMA', 'MASSI', 'NETO', 'NÁPOLES', 'OLIVEIRA', 'RODRIGUES', 'SANTOS', 'SILVA', 'SOARES']
    cargos = ['Estagiário', 'Assistente', 'Analista', 'Gerente', 'Diretor']
    salarios = np.random.randint(1000, 15000, n)
    
    start_date = pd.to_datetime('1990-01-01')
    end_date = pd.to_datetime('2005-01-01')
    
    dias = (end_date - start_date).days
    nascimentos_aleatorios = start_date + pd.to_timedelta(np.random.randint(0, dias, n), unit='D')
    nascimentos_formatados = nascimentos_aleatorios.strftime('%d/%m/%Y')    
    nomes_completos = [np.random.choice(nomes) + ' ' + np.random.choice(sobrenomes) for _ in range(n)]
    
    df = pd.DataFrame({'Nome': nomes_completos, 'Nascimento': nascimentos_formatados, 'Salário': salarios, 'Cargo': np.random.choice(cargos, n)})
    return df
```

```python
# Altera o salário de todos estagiários para 1000
def altera_salario_estagiarios(df):
    df.loc[df['Cargo'] == 'Estagiário', 'Salário'] = 1000
    return df
```

```python
df = gera_dataframe(100)
df = altera_salario_estagiarios(df)
df.head(n=20)
```

```python
#Duplicando a primeira linha
df = df.append(df.iloc[99], ignore_index=True)
# Criando uma linha com valores faltantes
df = df.append({'Nome': np.nan, 'Nascimento': '05/08/1990', 'Salário': 1000, 'Cargo': 'Estagiário'}, ignore_index=True)
df = df.append({'Nome': 'JÚLIO MASSI', 'Nascimento': np.nan, 'Salário': 1000, 'Cargo': 'Estagiário'}, ignore_index=True)
df = df.append({'Nome': 'MAYCON CRUZ', 'Nascimento': '21/04/1999', 'Salário': np.nan, 'Cargo': 'Estagiário'}, ignore_index=True)
df = df.append({'Nome': 'FILIPE CRUZ', 'Nascimento': '01/08/2001', 'Salário': np.nan, 'Cargo': 'Gerente'}, ignore_index=True)
# Criando uma linha com valores inválidos
df = df.append({'Nome': 'PEDRO NÁPOLES', 'Nascimento': '05/13/1990', 'Salário': 1000, 'Cargo': 'Gerente'}, ignore_index=True)
df = df.append({'Nome': 'OTÁVIO SOARES', 'Nascimento': '05/11/1790', 'Salário': 1000, 'Cargo': 'Asistente'}, ignore_index=True)
# Criando uma linha com outliers
df = df.append({'Nome': 'VICTOR RODRIGUES', 'Nascimento': '05/11/2990', 'Salário': 40000, 'Cargo': 'Direto'}, ignore_index=True)
df = df.append({'Nome': 'VANESSA OLIVEIRA', 'Nascimento': '105/11/1990', 'Salário': -5000, 'Cargo': 'Diretor'}, ignore_index=True)
df.tail(n=10)
```

```python
#Boxplot para identificar outliers
import matplotlib.pyplot as plt
df.boxplot(column='Salário')
plt.show()
```

```python
# Salvando o dataframe em um arquivo CSV
df.to_csv('clientesSujos.csv', index=False)
```
