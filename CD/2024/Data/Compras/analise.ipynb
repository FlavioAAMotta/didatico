```python
import pandas as pd
```

```python
df = pd.read_csv('dados_sinteticos.csv')
```

```python
df.head()
```

```python
# Excluindo células para criar um dataset com valores faltantes
# Primeiro, vamos excluir aleatoriamente 10% das células da coluna quantidade no dataset
import numpy as np
np.random.seed(0)
df.loc[df.sample(frac=0.1).index, 'quantidade'] = np.nan

# Agora, vamos excluir aleatoriamente 5% das células da coluna total no dataset
np.random.seed(1)
df.loc[df.sample(frac=0.05).index, 'total'] = np.nan
```

```python
df.head(n=50)
```

```python
# Vamos alterar valores de algumas células para criar um dataset com valores inconsistentes
# Primeiro, vamos alterar aleatoriamente 5% das células da coluna quantidade para valores negativos
np.random.seed(2)
df.loc[df.sample(frac=0.05).index, 'quantidade'] = -1

# Agora, vamos alterar aleatoriamente 5% das células da coluna hora para valores fora do intervalo permitido
np.random.seed(3)
df.loc[df.sample(frac=0.05).index, 'hora'] = np.random.choice([-1, 24, 25, 100], size=int(0.05*df.shape[0]))
```

```python
df.head(n=50)
```

```python
# Duplicando linhas para criar um dataset com valores duplicados
# Vamos duplicar aleatoriamente 5% das linhas do dataset
np.random.seed(4)
df = pd.concat([df, df.sample(frac=0.05)])
df = df.reset_index(drop=True)
```

```python
# Alterando algumas letras para criar um dataset com valores inconsistentes
# Vamos alterar aleatoriamente 5% das células da coluna status, apenas as letras 'a' e 'e' serão alteradas
np.random.seed(5)
# Selecionando aleatoriamente 5% das linhas para alterar
indices_to_change = df.sample(frac=0.05).index

# Função para substituir 'a' por 's' e 'e' por 'r'
def replace_letters(status):
    return status.replace('a', 's').replace('e', 'r')

# Aplicando a função de substituição apenas nos índices selecionados
for idx in indices_to_change:
    df.at[idx, 'status'] = replace_letters(df.at[idx, 'status'])
```

```python
df.head(n=50)
```

```python
# Imprimindo valores unicos da coluna status
print(df['status'].unique())
```

```python
#Alterando alguns modelos para suas versões de letras maiúsculas
# Vamos alterar aleatoriamente 5% das células da coluna modelo para suas versões em letras maiúsculas
np.random.seed(6)
# Selecionando aleatoriamente 0,5% das linhas para alterar
indices_to_change = df.sample(frac=0.0001).index

def to_upper_case(modelo):
    return modelo.upper()

# Aplicando a função de substituição apenas nos índices selecionados
for idx in indices_to_change:
    df.at[idx, 'modelo'] = to_upper_case(df.at[idx, 'modelo'])
print(indices_to_change)
```

```python
print(df['modelo'].unique())
```

```python
# Trocando alguns escritos Pneu por Peneu
# Vamos alterar aleatoriamente 15% das ocorrencias de Pneu por Peneu
np.random.seed(7)
# Selecionando aleatoriamente 15% das linhas para alterar
indices_to_change = df[df['modelo'].str.contains('Pneu')].sample(frac=0.15).index

def replace_Pneu(modelo):
    return modelo.replace('Pneu', 'Peneu')

# Aplicando a função de substituição apenas nos índices selecionados
for idx in indices_to_change:
    df.at[idx, 'modelo'] = replace_Pneu(df.at[idx, 'modelo'])

print(indices_to_change)
print(df[df['modelo'].str.contains('Peneu')].shape[0])
```

```python
# Salvando o dataset com valores faltantes, inconsistentes, duplicados e incorretos
df.to_csv('dados_sinteticos_com_erros.csv', index=False)
```
