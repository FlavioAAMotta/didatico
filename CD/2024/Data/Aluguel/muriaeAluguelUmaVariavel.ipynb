```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo a semente para reprodutibilidade
np.random.seed(42)

# Número de linhas
n_linhas = 500

# Gerando dados sintéticos
data = {
    'Area': np.random.randint(30, 300, n_linhas),  # Área em metros quadrados
}

```

```python
# DataFrame sem ruído
df_sem_ruido = pd.DataFrame(data)
base_price = 800
df_sem_ruido['Preco'] = (base_price +
                         df_sem_ruido['Area'] * 20)

# Adicionando ruído ao DataFrame sem ruído para criar um com ruído
ruído = np.random.normal(0, 250, n_linhas)  # Média 0, Desvio padrão 250
df_com_ruido = df_sem_ruido.copy()
df_com_ruido['Preco'] += ruído  # Adicionando ruído
```

```python

# Plotando os dados
plt.figure(figsize=(10, 6))
plt.scatter(df_sem_ruido['Area'], df_sem_ruido['Preco'], alpha=0.5, label='Sem ruído', color='b')
plt.scatter(df_com_ruido['Area'], df_com_ruido['Preco'], alpha=0.5, label='Com ruído', color='r')
plt.title('Preço do Aluguel x Área da Casa')
plt.xlabel('Área (m²)')
plt.ylabel('Preço do Aluguel (R$)')
plt.legend()
plt.show()
```

```python
# Salvando os datasets em arquivos CSV de nome 'aluguel_uma_variavel_sem_ruido.csv' e 'aluguel_uma_variavel_com_ruido.csv'
df_sem_ruido.to_csv('aluguel_uma_variavel_sem_ruido.csv', index=False)
df_com_ruido.to_csv('aluguel_uma_variavel_com_ruido.csv', index=False)
```
